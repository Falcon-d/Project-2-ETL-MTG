```python
# Module used to connect Python with MongoDb
import pymongo
from pymongo import MongoClient
import pandas as pd
```

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the mtg database in Mongo
db = client.mtg
```

```python
# Query all rows
# Here, db.raw_count refers to the collection 'raw_count'
raw_count = db.raw_count.find()
mtg_merged_df = db.mtg_merged_df.find()
```

```python
# Iterate through each row in the collection
for row in raw_count:
    print(row)
    
for row in mtg_merged_df:
    print(row)
```

```python
#select the collection within the database
collection = db.mtg_merged_df

#convert entire collection to Pandas dataframe
df = pd.DataFrame(list(collection.find()))
df
```

```python
#select the collection within the database
rcollection = db.raw_count

df = pd.DataFrame(list(rcollection.find()))
df
```

```python
# Number of cards that rank 0-7 for Converted Mana Cost 
df.groupby('Converted Mana Cost').count()
```

```python
# Number of cards that rank 0-7 for Power
df.groupby('Power').count()
```

```python
# Number of cards that rank 0-7 for Toughness
df.groupby('Toughness').count()
```

```python
# Number of cards that rank 0-7 for Type Line
df.groupby('Type Line').count()
```
